```python
# Redes neuronales(RN), son de gran peso   
# MV = Maquina virtual
# Entre más recurso gaste mas gasto :(
# Imagen es plus de RedNeuronal
# Cada imagen esta compuesta por colores, no es una imagen.
# Colores estan compuesto por triada de números
# RN no toma la imagen cuadrada  
# rN son sistema matematico de alto volumen de variables donde va a intentar asignar un peso

# Cuando llega una imagen
# sistema de reducción de dimensiones
# Pasarlo a blanco y negro
# Las RN trabajon con blanco y negro
# RN convolucional si trabaja con color
# convolucion es una operacion matematica, se usa como una forma de utuliza filtros de señales analogas
# 
```

```python
import cv2
import numpy as np
import imutils
from google.colab.patches import cv2_imshow
imagen = cv2.imread('/content/drive/MyDrive/Colab Notebooks/19-11-2.jpg')
```

```python
cv2_imshow(imagen)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
rojoBajo1 = np.array([10, 40, 90], np.uint8)
rojoAlto1 = np.array([200, 155, 155], np.uint8)
```

```python
image = imutils.resize(imagen, width=320)
cv2_imshow(image)
print(image[0][0])
```

```python
imagegray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(imagegray)
print(imagegray[0][0])
```

```python
imagegray = cv2.cvtColor(imagegray,cv2.COLOR_GRAY2BGR)
cv2_imshow(imagegray)
print(imagegray[0][0])
```

```python
imageHSV = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2_imshow(imageHSV)
print(imageHSV[0][0])
```

```python
maskRojo1 = cv2.inRange(imageHSV, rojoBajo1, rojoAlto1)
#mask = cv2.add(maskRojo1,maskRojo2)
mask = cv2.medianBlur(maskRojo1, 7)

cv2_imshow(mask)
```

```python
redDetected = cv2.bitwise_and(image,image,mask=mask)
cv2_imshow(redDetected)
```

```python
invMask = cv2.bitwise_not(mask)
bgGray = cv2.bitwise_and(imagegray,imagegray,mask=invMask)
cv2_imshow(bgGray)
```

```python
finalImage = cv2.add(bgGray,redDetected)
# Visualización
cv2_imshow(finalImage)
```
